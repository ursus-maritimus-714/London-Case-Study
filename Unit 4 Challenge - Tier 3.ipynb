Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#Having a look at the columns situation in the current DF. We want to move to a DF structure 
#where there's a single column with all of the datetimes across the current column headings, and each row will be an observation
properties_transposed.columns
```

```python
#Collapsed datetimes column now needs a new, descriptive name, as does current value column
melted_properties = melted_properties.rename(columns = {0: 'Month', 'value': 'Mean_price'})
```

```python
#Verifying our renames
melted_properties.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
#Changing mean price column to float (is "coerce" argument correct here to deal with non-parsable strings?")
melted_properties['Mean_price'] = pd.to_numeric(melted_properties['Mean_price'], errors='coerce')
```

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
#Using unique() method to first assess what's in our 'Borogh' column (i.e are we dealing with just actual Boroughs?)
melted_properties['Borough'].unique()
```

```python
#Removing all records for the remaining non-Boroughs from the DF 
#(I struggled with this piece beyond building the list of the non-Boroughs part, and relied on lower tier for the clever negation method. 
#Which I'm sure I'll use again!)
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']
```

```python
#Filtering on the condition that the rows' values for London_Borough is in the nonBoroughs list.
No_NanDF[No_NanDF.Borough.isin(nonBoroughs)]
```

```python
#Negation operator trick to obtain only those rows whose values for Borough that are NOT in the nonBoroughs list
No_NanDF[~No_NanDF.Borough.isin(nonBoroughs)]
```

```python
#Reassign using the negation operator trick and look at subset of the data to verify nn-Borough removal
No_NanDF = No_NanDF[~No_NanDF.Borough.isin(nonBoroughs)]
```

```python
#Index_reset on mpypb
mpypb = mpypb.reset_index()
mpypb.head(25)
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#Fxn to generate 2018/1998 price ratio
def p_ratio(x):
    price1998 = float(x['Mean_price'][x['Year']==1998])
    price2018 = float(x['Mean_price'][x['Year']==2018])
    p_rat = [price2018/price1998]
    return p_rat
```

```python
#Run on the same test borough as previously (City of London)
p_ratio(mpypb[mpypb['Borough']=='City of London'])
```

```python
#Create empty dictionary to house ratio for all London boroughs
all_ratios = dict()
```

```python
#Iterates through all uniques in Borough Column of DF and generates 2018/1998 price ratio for each
for y in mpypb['Borough'].unique():
    brgh = mpypb[mpypb['Borough'] == y]
    all_ratios[y] = p_ratio(brgh)
print(all_ratios)
```

```python
#Convert dictionary of unique ratios to a DF
all_ratios_df = pd.DataFrame(all_ratios)
```

```python
all_ratios_df.head(10)
```

```python
#As with the original DataFrame, we want to transpose this and reset the index
all_ratios_df_transposed = all_ratios_df.T
all_ratios_df = all_ratios_df_transposed.reset_index()
all_ratios_df.head(20)
```

```python
#And finally to rename the columns descriptively
all_ratios_df.rename(columns={'index':'Borough', 0: '2018_1998_ratio'}, inplace=True)
all_ratios_df.head(25)
```

```python
#Look at the 32 Borough historical ratios in descending order
print(all_ratios_df.sort_values(by='2018_1998_ratio', ascending=False))
```

```python
#Boroughs plot by decreasing price ratio 2018/1998
dec_ratio = all_ratios_df.sort_values(by='2018_1998_ratio', ascending=False)
```

```python
#Bar plot of ratios in desending order
bp = dec_ratio[['Borough','2018_1998_ratio']].plot(kind='bar')
bp.set_xticklabels(dec_ratio.Borough)
```
